```python
#import libraries
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
import yfinance as yf 
import pandas_datareader as web
```

```python
# read csv for top list
stock_list = pd.read_csv("sandp100.csv")
stock_list.head()
```

```python
#extract stock symbols
tickers = stock_list['Symbol']
tickers
```

```python
#replace some values that are presented different in yahoo
stock_list['Symbol'] = stock_list['Symbol'].replace(['BRK.B'],'BRK-B')
#extract symbols into separate list
ticker_list = stock_list['Symbol'].tolist() 
ticker_list
```

```python
# Define the ticker symbol
ticker_symbol = ticker_list

stock_historical_data = pd.DataFrame()

for ticker_symbol in ticker_list: 
    # Create a Ticker object
    ticker = yf.Ticker(ticker_symbol)

    # Fetch historical market data for the last 30 days
    historical_data = ticker.history(start= '2022-01-01', end= '2022-12-31', auto_adjust=False, actions=False)  # data for the last month

    historical_data['Ticker'] = ticker_symbol

    
    stock_historical_data = pd.concat([stock_historical_data, historical_data])


    # Display a summary of the fetched data
    print(f"Summary of Historical Data for {ticker_symbol}:")
    print(historical_data[['Adj Close','Close','Open', 'High', 'Low', 'Volume']])
    print("\n")





stock_historical_data.reset_index(inplace=True)

stock_historical_data.to_csv("C:/Users/user/Downloads/AI Siminar/AI-Seminar-Project/Finance/Stock_historicaldata_2022.csv", index=False)

```

```python

# Sort the DataFrame by the 'Date' column in ascending order
stock_sortedbydate = stock.sort_values(by='Date', ascending=True)

# Reset the index to create sequential IDs (0, 1, 2, 3, ...)
stock_sortedbydate.reset_index(drop=True, inplace=True)
stock_sortedbydate
```

```python
#adjust pricing formating to 2 decimal points
pd.options.display.float_format = '{:,.2f}'.format
print(stock_sortedbydate)
print(stock_sortedbydate.head())
```

```python
#adjust the volume column to remove extra decimals
stock_sortedbydate['Volume']=stock_sortedbydate['Volume'].astype(int).apply(lambda x : "{:,}".format(x))
stock_sortedbydate.head()
```

```python
stock_df= stock_sortedbydate.rename(columns={'Ticker': 'Symbol', 'Adj Close': 'Adj_Close'})
stock_df.head()
```

```python
# merge with original s&p100 list for comprehensive 
stock_data_2021= pd.merge(stock_list, stock_df, on='Symbol',  how='left')
stock_data_2021.tail()
```
